```python
df = pd.read_csv('C:\\Users\\harshit.vashistha\\Desktop\\Recommendation project\\foodsmartz\\recommendation\\dataset\\recommendation_data')
```

```python
df1 = df.iloc[:1000, :]
```

```python
df1.drop(columns=['Unnamed: 0'], inplace= True)
```

```python
df1['combined'] = df1['cuisines'] + " " + df1['dish_liked']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df1['combined'])
```

```python
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
df1.head()
```

```python
def recommend_restaurants(restaurant_name, cosine_sim=cosine_sim):
    idx = df1.index[df1['name'] == restaurant_name][0]

    sim_scores = list(enumerate(cosine_sim[idx]))

    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    sim_scores = sim_scores[1:5]

    restaurant_indices = [i[0] for i in sim_scores]
    
    return df1.iloc[restaurant_indices]
```

```python
recommended = recommend_restaurants('Jalsa')
```
